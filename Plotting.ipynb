```python
import pandas as pd
from matplotlib import pyplot as plt
import matplotlib.patches as mpatches
import numpy as np
```

# Every 5th Section Plotting

```python
mg49_5 = pd.read_csv('//ibs9010/current_data/Data_Keshav/MG49stats_every5th_Section1to100.csv')
mg48_5 = pd.read_csv('//ibs9010/current_data/Data_Keshav/MG48stats_every5th_Section1to100.csv')
```

```python
mg48_5.head(60)
```

```python
mg48_5.describe()
```

```python
mg49_5.describe()
```

```python
# a scatter plot 
mg48_5.plot(kind='scatter',x='slice_number',y='TP_per')

plt.show()
```

```python
ax = plt.gca()

#plt.figure(figsize = (50,50))
mg49_5.plot(kind='line',x='slice_number',y='TP_per',color='green',ax=ax,figsize = (20,10),legend='best')
mg48_5.plot(kind='line',x='slice_number',y='TP_per', color='red',linewidth=2, ax=ax,)


#mg49_5.plot(kind='line',x='slice_number',y='FP_per', color='green',linewidth=2, ax=ax)
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left',prop={'size': 20})
plt.xlabel('Slice', fontsize=18)
plt.ylabel('True Positive (%)', fontsize=16)
ax.tick_params(axis='x', which='major', labelsize=15)
ax.tick_params(axis='y', which='minor', labelsize=8)

max_value = mg49_5['slice_number'].max()
min_value = mg49_5['slice_number'].min()
number_of_steps = 5
l = np.arange(min_value, max_value+1, number_of_steps)
ax.set(xticks=l, xticklabels=l)
ax.set(yticks=l, yticklabels=l)

red_patch = mpatches.Patch(color='red', label='MG48_TP%')
green_patch = mpatches.Patch(color='green', label='MG49_TP%')
#black_patch= mpatches.Patch(color='black', label='FP%')
plt.legend(handles=[red_patch,green_patch])

plt.title("MG48 and MG49 Inference - True Positive from Every 5th Section")

plt.show()
#plt.savefig('//ibs9010/current_data/Data_Keshav/MG48and49_TP_5thsection.png',dpi=300,transparent=True)
```

```python
# MG49 Full plotting

plt.rc('figure', figsize=(12, 6))   # this is to overwrite default aspect of graph to make x-axis longer

fig, ax1 = plt.subplots()

l1  =['TP%','FN%','FP%']

# data to be plotted from dataframe - temperatures as line graphs
#Dates = doystats['Fdate']
slices = mg49_5['slice_number']
TP_per = mg49_5['TP_per']
FN_per = mg49_5['FN_per']
FP_per = mg49_5['FP_per']

ax1.plot(slices, TP_per, 'g', slices, FN_per, 'r', slices, FP_per, 'k')

# Make the y-axis label
ax1.set_xlabel('Slice Number')
ax1.set_ylabel('Percentage')
ax1.tick_params('y')
ax1.set_ylim(top=150)
ax1.set_ylim(bottom=0)#-50
# Creating twin axes for precipitation as a bar graph on secondary XY axes
ax2 = ax1.twiny()
ax3 = ax2.twinx()

#data for bar graph
#slice_number = mg49_5['slice_number']
#TP = mg49_5['TP']
#ax3.bar(slice_number, TP_per, color='b')
#ax2.set_xticks([])                # to hide ticks on the second X-axis - top of the graph
#ax3.set_ylabel('True Positive (absolute numbers)', color='b')
#ax3.tick_params('y', colors='b')
#ax3.set_ylim(top=800)

# Formatting X-axis to be evernly spaced
max_value = mg49_5['slice_number'].max()
min_value = mg49_5['slice_number'].min()
number_of_steps = 2
l = np.arange(min_value, max_value+1, number_of_steps)

ax.set(xticks=l, xticklabels=l)

red_patch = mpatches.Patch(color='red', label='FN%')
green_patch = mpatches.Patch(color='green', label='TP%')
black_patch= mpatches.Patch(color='black', label='FP%')
plt.legend(handles=[red_patch,green_patch,black_patch])

plt.title("MG49 Inference - Data From Every 5th Section of MG48")
ax1.grid(True)
fig.tight_layout()
# Saving graph to file
plt.savefig('//ibs9010/current_data/Data_Keshav/MG49_every5thsection.png',dpi=300,transparent=True)
```
